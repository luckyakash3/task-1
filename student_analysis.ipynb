```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student-mat.csv', sep=';')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Average Final Grade
avg_grade = df['G3'].mean()

print("Average Final Grade =", avg_grade)
```

```python
# Students Scoring Above 15
students_above_15 = df[df['G3'] > 15]

print("Number of students scoring above 15 =", len(students_above_15))
```

```python
#Correlation Between Study Time and Grades
correlation = df['studytime'].corr(df['G3'])

print(correlation)
```

```python
#Gender Performance Analysis
gender_avg = df.groupby('sex')['G3'].mean()

print(gender_avg)
```

```python
# Visualization 1 (Histogram)
plt.figure(figsize=(8,5))

plt.hist(df['G3'], bins=10)

plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Number of Students')

plt.show()
```

```python
# Visualization 2 (Scatter Plot)
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='studytime',
    y='G3',
    data=df
)

plt.title('Study Time vs Final Grade')

plt.show()
```

```python
# Visualization 3 (Bar Chart)
gender_avg.plot(kind='bar')

plt.title('Average Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Grade')

plt.show()
```

CONCLUSION

1. Average final grade was calculated using G3 column.
2. Students scoring above 15 were identified.
3. Study time and grades showed a positive relationship.
4. Gender-wise performance comparison was performed.
5. Visualizations provided insights into student performance patterns.

The analysis indicates that students who spend more time studying generally achieve better academic results.
